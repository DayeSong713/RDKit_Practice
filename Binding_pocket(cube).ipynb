```python
with open('1t46.pdb','r') as f:
    ATOM_coord=[]
    while True:
        line=f.readline()
        if line[0:4]=='ATOM':
            x = float(line[30:38]) # float 8
            y = float(line[38:46]) # float 9
            z = float(line[46:54])
            coord=[x,y,z]
            ATOM_coord.append(coord)
        if not line:
            break
    print(ATOM_coord)
```

```python
with open('1t46.pdb','r') as f:
    STI_coord_x=[]
    STI_coord_y=[]
    STI_coord_z=[]
    
    while True:
        line=f.readline()
        if line[0:6]=='HETATM':
            resName = line[17:20].strip()
            if resName == 'STI':
                x = float(line[30:38]) # float 8
                y = float(line[38:46]) # float 9
                z = float(line[46:54])
                STI_coord_x.append(x)
                STI_coord_y.append(y)
                STI_coord_z.append(z)
        if not line:
            break
    x=sum(STI_coord_x)/len(STI_coord_x)
    y=sum(STI_coord_y)/len(STI_coord_y)
    z=sum(STI_coord_z)/len(STI_coord_z)
    center=[x,y,z]
    print(center)
```

```python
range_x=[center[0]-12,center[0]+12]
range_y=[center[1]-12,center[1]+12]
range_z=[center[2]-12,center[2]+12]
print(range_x,range_y,range_z)
range_coord=[range_x,range_y,range_z]
print(range_coord)
```

```python
ATOM_index=[]
for i in range(len(ATOM_coord)):
    a=0
    for j in range(len(ATOM_coord[i])):
        if (ATOM_coord[i][j]>range_coord[j][0])&(ATOM_coord[i][j]<range_coord[j][1]):
            a+=1
    if a==3:
        ATOM_index.append(i)
```

```python
print(ATOM_index)
```

```python
for i in range(len(ATOM)):
    if i in ATOM_index:
        print(ATOM[i])
```

```python
Close_residues=[]
for i in range(len(ATOM)):
    if i in ATOM_index:
        Close_residues.append(ATOM[i-1][21:26])
print(Close_residues)
```

```python
Close_residues=set(Close_residues)
```

```python
ATOM2=[]
for i in range(len(ATOM)):
    if ATOM[i][21:26] in Close_residues:
        ATOM2.append(ATOM[i])
print(ATOM2)
```

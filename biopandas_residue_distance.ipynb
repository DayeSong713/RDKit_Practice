```python
from math import *
```

```python
def euclidean_distance(x, y):
    return sqrt(sum(pow(a - b, 2) for a, b in zip(x, y)))

print(euclidean_distance([31.938,29.660,49.840], [31.424,32.568,47.447]))
#Euclidean distance 구하는 시용자 함수를 만든다.
#첫 번째 행과 두 번째 행의 좌표를 대입해 보았을 때 정상적으로 작동한다.
```

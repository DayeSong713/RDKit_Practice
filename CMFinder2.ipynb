```python
from biopandas.pdb import PandasPdb
from math import *
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
pdb_1u2w_1=PandasPdb().read_pdb('/home/ailon/Desktop/python_practice/PDBfile/1U2W.pdb')
```

```python
pdb_1u2w_2=pdb_1u2w_1.df['ATOM']
```

```python
pdb_1u2w_4=pdb_1u2w_3[(pdb_1u2w_3['atom_name']=='CB')&(((pdb_1u2w_3['alt_loc']=='')|(pdb_1u2w_3['alt_loc']=='A')))|(pdb_1u2w_3['residue_name']=='GLY')&(pdb_1u2w_3['atom_name']=='CA')&((pdb_1u2w_3['alt_loc']=='')|(pdb_1u2w_3['alt_loc']=='A'))]
```

```python
pdb_1u2w_4
```

```python
pdb_1u2w_5=pdb_1u2w_4.loc[:,['residue_name','residue_number','x_coord','y_coord','z_coord']]
```

```python
pdb_1u2w_5
```

```python
pdb_1u2w_5.loc[:,['x_coord','y_coord','z_coord']]
```

```python
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2) for a,b in zip(x,y)))
```

```python
arr1=np.empty((len(pdb_1u2w_5),len(pdb_1u2w_5)))
```

```python
for i in range(len(pdb_1u2w_5)):
    for j in range(len(pdb_1u2w_5)):
        residue_pair_1=pdb_1u2w_5.iloc[i][['x_coord','y_coord','z_coord']].values
        residue_pair_2=pdb_1u2w_5.iloc[j][['x_coord','y_coord','z_coord']].values
        arr1[i][j]=euclidean_distance(residue_pair_1,residue_pair_2)
```

```python
arr1
```

```python
plt.imshow(arr1,cmap='bone')
```

```python
arr2=np.empty((len(pdb_1u2w_5),len(pdb_1u2w_5)))
```

```python
for i in range(len(pdb_1u2w_5)):
    for j in range(len(pdb_1u2w_5)):
        if arr1[i][j]<=8:
            arr2[i][j]=1
        else:
            arr2[i][j]=0
```

```python
arr2
```

```python
plt.imshow(arr2,cmap='bone')
```
